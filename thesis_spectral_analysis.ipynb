```python
from __future__ import print_function

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import json
import os
import mne

import ecogtools
from scipy.stats import ttest_ind

%matplotlib inline
```

```python
mne.set_log_level(verbose=False)
```

```python
TR_times = {"tmin":-2., "tmax":1.}
TR_baseline = (0., 1.)
```

```python
freqs = np.linspace(80, 150, 50)
n_cycles = freqs/2.0
time_bandwidth = 8.0
```

```python
patient_num1 = "2002"
channels1 = ['RTG22', 'RTG23', 'RTG24', 'RTG30', 'RTG31', 'RTG32', 'RTG38', 'RTG39', 'RTG40', 'RTG46', 'RTG47', 'RTG48']
```

```python
data1 = ecogtools.ToM_Localizer(patient_num1)

# Get the times for task from behavioral data1
tmin = int(data1.trig_and_behav.loc[0, 'trigger_time']-10)
tmax = int(data1.trig_and_behav.loc[59, 'trigger_time']+10)

# Get only physiology data1 that we actually want to look at for task.
data1.phys.crop(tmin=tmin, tmax=tmax)

# Load data1 and pick only channels that we care about.
data1.phys.load_data()
data1.phys.pick_channels(channels1)

data1.initialize_epochs_object(channels1, tmin=-4., tmax=2., baseline=None)

# Calculate power for Time of Response
belief_tr_power1 = mne.time_frequency.tfr_multitaper(data1.epochs['b/time_of_resp'], freqs=freqs, n_cycles=n_cycles,
                       time_bandwidth=time_bandwidth, return_itc=False)
photo_tr_power1 = mne.time_frequency.tfr_multitaper(data1.epochs['p/time_of_resp'], freqs=freqs, n_cycles=n_cycles,
                       time_bandwidth=time_bandwidth, return_itc=False)

TR_power1 = data1.compute_diff_power(belief_tr_power1, photo_tr_power1)
```
